```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
import GrafOpt as go
```

```python
x,y=np.loadtxt('C:/Users/Yovanni/Downloads/Regresion lineal/IR1.txt',unpack=True);

G=go.RL(x,y);
print(G);

xi=x[0];
xf=x[-1];

X=np.zeros(101);
Y=np.zeros(101);
dx=(xf-xi)/100;

for i in range(0,101):
    X[i]=xi+i*dx;
    Y[i]=go.f(X[i],G[0],G[1]);

fig, ax = plt.subplots();
ax.plot(x,y,'co');
ax.plot(X,Y,'r-');
ax.set_title('Gr치fica del 치ngulo de reflexi칩n, interfaz aire-vidrio')
L=[r'$Observaci칩n$', r'$Ajuste \ lineal: \theta_i=\theta_r$']
ax.legend(L)
ax.set_xlabel(r"$\theta_r \ (^\circ)$", fontsize=15, fontname='Times New Roman')
ax.set_ylabel(r"$\theta_i \ (^\circ)$", fontsize=15, fontname='Times New Roman')
plt.grid();
plt.show()
```
